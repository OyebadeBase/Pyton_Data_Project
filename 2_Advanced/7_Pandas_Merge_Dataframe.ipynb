```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\P R O B O O K\Downloads\data_jobs.csv")
df = data.copy()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
df_US_pivot=df_US_pivot=df_US.pivot_table(columns='job_title_short', index='job_posted_month', aggfunc='size')
df_US_pivot=df_US_pivot.reset_index()
df_US_pivot['job_posted_month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format = '%B').dt.month
df_US_pivot=df_US_pivot.sort_values('job_posted_month_no')
df_US_pivot=df_US_pivot.set_index('job_posted_month')
df_US_pivot=df_US_pivot.drop(columns='job_posted_month_no')

df_US_pivot
```

```python
df_US_software_pivot = pd.read_csv(r'C:\Users\P R O B O O K\Downloads\software_jobs (1).csv', index_col='job_posted_month')
df_US_software_pivot
```

```python
df_US_merged= df_US_pivot.merge(df_US_software_pivot, on= 'job_posted_month')
```

```python
top_5 = (df_US_merged.
         sum().
         sort_values(ascending=False).
         head().
         index
)
top_5
```

```python
df_US_merged[top_5].plot(kind='line')
plt.title('Monthly Job Postings for Top Tech Job in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0,20000)
plt.legend()
plt.show()
```
