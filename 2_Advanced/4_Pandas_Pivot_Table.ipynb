```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\P R O B O O K\Downloads\data_jobs.csv")
df = data.copy()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.pivot_table(values='salary_year_avg',columns='job_title_short', index='job_country',aggfunc='median')
```

```python
top_countirs = df['job_country'].value_counts().head(6).index

df_job_country_salary=df.pivot_table(values='salary_year_avg',
               columns='job_title_short',
                 index='job_country',
                 aggfunc='median'
)

df_job_country_salary = df_job_country_salary.loc[top_countirs]

job_title = ['Data Analyst', 'Data Engineer', 'Data Scientist']

df_job_country_salary = df_job_country_salary[job_title]

df_job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary ($USD)')
plt.xlabel('')
plt.title('Median Salary By Country and Job Title')
plt.xticks(rotation= 45, ha = 'right')
plt.show()
```
